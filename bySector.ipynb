```python
import pandas as pd
```

```python
import csv

with open('data/nasdaq-companies.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=';')
    ticker = []
    name = []
    sector = []
    for row in readCSV:
        ticker.append(row[0])
        name.append(row[1])
        sector.append(row[6])
        
with open('data/securities.csv') as basecsvfile:
    basereadCSV = csv.reader(basecsvfile, delimiter=',')
    for row in basereadCSV:
        if row[0] not in ticker:
            ticker.append(row[0])
            name.append(row[1])
            sector.append('')

d = {'Ticker' : ticker, 
    'Company' : name,
    'Sector' : sector}
        
df = pd.DataFrame(data=d)

df = df.drop([0])

df.sort_values(by='Ticker')

df
```

```python
df = df.sort_values(by='Ticker', ascending=True)

df
```

```python
df.to_csv('Companies By Sector.csv', index=False)
```
